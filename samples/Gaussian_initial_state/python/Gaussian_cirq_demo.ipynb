Implementation of an algorithm to prepare a single-variable Gaussian state on a quantum computer, as described in the paper by A. Kitaev and W. Webb: https://arxiv.org/abs/0801.0342. The algorithm below is slightly modified from the original, and replaces the supplementary memory qubits with N-qubit unitary transformations. The implementation is performed in pyQuil.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from gauss_wavefcn import GaussianWavefunction
```

Specify the number of qubits to be used, as well as the standard deviation and mean of the Gaussian wavefunction

```python
n_qubits = 7
std_dev = (2**n_qubits)/6.  # pick a value that satisfies the limit described in paper
mean = 2**(n_qubits-1) - 0.5  # place the mean in the center
```

Prepare the Gaussian wavefunction

```python
wavefunction = GaussianWavefunction(sigma_=std_dev, mu_=mean, num_qubits=n_qubits).gaussian_wavefunc()
```

Plot its probability amplitudes

```python
# get final state vector
state_vector = wavefunction.state_vector()
# get amplitudes
amplitudes = [i**2 for i in state_vector]
# rounding
amplitudes_round = np.around(amplitudes, 3)
# print rounded amplitudes
print(amplitudes_round)
# plot amplitudes
plt.plot(amplitudes)
plt.savefig('wavefunction.png')
plt.show()
```
